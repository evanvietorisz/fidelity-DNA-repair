```python
###using probability for rates
#the fast (left walker's) rate is now the base rate


import numpy as np
import random #why can't grab the one func I need
import math ##TODO only get necessary function(s)
from fractions import Fraction
import matplotlib.pyplot as plt

class system:
    
    def __init__(self, L):
        self.Lbound = 0
        self.Rbound = L-1
        self.status = 'active'
        self.lifetime = 0
        self.initial_length = L
        
    def display(self):
        print("Lbound =",self.Lbound)
        print("Rbound =",self.Rbound)
        print("status =",self.status)
        print("lifetime =",self.lifetime,"\n")
        
    def Lstep(self,p,inf_5prime):
        #step left walker
        if inf_5prime == False and self.Lbound == 0:
            self.Lbound += 1
        else:
            if np.random.random_sample() < p:
                self.Lbound -= 1
            else:
                self.Lbound += 1            
    
    def Rstep(self,p,r,hold_3prime):
        #step right walker
        #extra probabilty gate makes scales down the right walker's rate on average
        #only supports r >= 1
        if hold_3prime == False and np.random.random_sample() < 1/r:
            if self.Rbound == self.initial_length-1:
                self.Rbound -= 1
            else:
                if np.random.random_sample() < p:
                    self.Rbound += 1
                else:
                    self.Rbound -= 1

class curve:
    
    def __init__(self,y_vals,x_vals,capture_time):
        self.y_vals = y_vals
        self.x_vals = x_vals
        self.capture_time = capture_time
                    
                    
class result:
    
    def __init__(self,ensemble,N,L,p,r,inf_5prime,delay_3prime,tmax,eval_times,padding,
                 remaining_systems,avg_half_life,curves):
        #parameters
        self.ensemble = ensemble
        self.N = N
        self.L = L
        self.p = p
        self.r = r
        self.inf_5prime = inf_5prime
        self.delay_3prime = delay_3prime
        self.tmax = tmax
        self.eval_times = eval_times
        self.padding = padding
        
        #caculated paramters
        self.extremal_Lbound =  self.extremal_Lbound()
        self.extremal_Rbound = self.extremal_Rbound()
        
        #experimental results
        self.remaining_systems = remaining_systems
        self.avg_half_life = avg_half_life
        self.curves = curves
        
    def extremal_Lbound(self):
        extremal_Lbound = self.L
        for sys in self.ensemble:
            if sys.Lbound < extremal_Lbound:
                extremal_Lbound = sys.Lbound
            return(extremal_Lbound)
    
    def extremal_Rbound(self):
        extremal_Rbound = 0
        for sys in self.ensemble:
            if sys.Lbound > extremal_Rbound:
                extremal_Rbound = sys.Rbound
            return(extremal_Rbound)
    
    def display_params(self):
        print('INPUT PARAMETERS')
        print('number of systems: ',N)
        print('initial length of systems: ',L)
        print('outward probability: ',p)
        print("ratio of rate going towards 3' end over 5' end: ",r)
        print("infinite 5' end: ",inf_5prime)
        print("delay on 3' end: ",delay_3prime)
        print('number of timesteps: ',tmax,'\n')
    
    def display_results(self):
        print('RESULTS')
        print('extremal left boundary after experiment: ',self.extremal_Lbound)
        print('extremal right boundary after experiment: ',self.extremal_Rbound)
        print('proportion of remaining systems after experiment: ',self.remaining_systems[tmax-1])
        print('average half-life of a system: ',self.avg_half_life,'\n')
        
    def summarize(self):
        #bring in quantities
        remaining_systems = self.remaining_systems
        avg_half_life = self.avg_half_life
        curves = self.curves
        
        
        fig, axs = plt.subplots(1,2,figsize=(10,5))
        #remaining systems w time
        axs[0].plot(remaining_systems)
        axs[0].set(xlim=([0,len(remaining_systems)]),ylim=([0,1.1]),xlabel="time (au)", 
                   title="Fraction of remaining systems \n (Average half-life: %.2f steps)" % avg_half_life)

        for curve in curves:
            axs[1].plot(curve.x_vals,curve.y_vals,label = "time = %i steps" % curve.capture_time)
        axs[1].set(xlabel="position (bp)",title="Probability of Occupation")
        axs[1].legend()

        plt.show()

        return()
    
def experiment(N,L,p,r,inf_5prime,delay_3prime,tmax,eval_times,padding):
    #output is result object
    
    remaining_systems = []
    avg_half_life = 0
    curves = []
    
    #make ensemble
    ensemble = []
    for i in range(N):
        ensemble.append(system(L))


    #advance through time, using time_step() as appropriate
    for t in range(tmax):
        
        #boolean about whether to fix the 3' end
        hold_3prime = t < delay_3prime
        #tally to keep track of nuber of systems remaining
        remaining = 0
        
        #advance systems
        for sys in ensemble:
            if sys.status == 'active':
                sys.Lstep(p,inf_5prime)
                sys.Rstep(p,r,hold_3prime)  
                
                #screen for death
                if sys.Lbound >= sys.Rbound:
                    sys.status = 'inactive'
                    
                #if still alive, increment lifetime and tally 
                if sys.status == 'active':
                    sys.lifetime += 1
                    remaining += 1
        
        #record fraction of systems still alive after timestep
        remaining_systems.append(remaining/N)
                
        #capture curve at desired eval_times
        # t+1 allows it to capture the 'tmax' timestep
        if t+1 in eval_times:
            
            #make graphical representation
            y_vals = [0] * (L + padding)
            x_vals = list(range(-padding,L))
            for sys in ensemble:
                for ind,val in enumerate(x_vals):
                    if sys.Lbound <= val <= sys.Rbound:
                        y_vals[ind] += 1
            #scale by N
            for ind,val in enumerate(y_vals):
                y_vals[ind] = y_vals[ind]/N
                
            curves.append(curve(y_vals,x_vals,t+1))
            
    #calculate average half-life
    for sys in ensemble:
        avg_half_life += sys.lifetime
    avg_half_life = avg_half_life/(2*N)
    
    #create result object
    res = result(ensemble,N,L,p,r,inf_5prime,delay_3prime,tmax,eval_times,padding,
                 remaining_systems,avg_half_life,curves)
    
    return(res)


###Program

#basic parameters
N = 1000
L = 80
p = .51
tmax = 1000

#correction parameters
r = 1
inf_5prime = False
padding = 0
delay_3prime = 0

#plotting
eval_times = [100,500,1000]

#actions
result = experiment(N,L,p,r,inf_5prime,delay_3prime,tmax,eval_times,padding)
result.display_params()
result.display_results()
result.summarize()

###
    
#Instructions
#L is length of system (int)
#N is number of systems (int)
#p is probability of a random walker taking a step out (float (0,1))
#r is ratio of rate going towards 3' end over rate going towards 5' end (float)
    #(ratio of distances covered by a step Right vs a step left)
#tmax is the number of timesteps (int)
    #(program *reaches* time tmax, not tmax-1)
#inf_5prime is whether or not the 5' end is infinite (bool)
#padding is the number of base pairs you alot in the negative direction to visualize infinite 5' end (int)
#delay_3prime is the number of steps of the faster walker the system waits to move the 3' end (int)

#TODO:
#specify functions in import
#clean up summarize() -- currently takes in a lot of parameters
#make portrait object?
#make cleaning function for data where new information shows up in the occupancy plot every two bp
#add feature to add bp on 3' end to mimic polymerase
#make output figures contain the N,L,p,r,etc parameters for reference
#make it so that delays on 3' end don't cause ugly downward lines on occupancy plot



```
